## Step 1: Load the Iris Dataset

We will use `sklearn.datasets.load_iris` to load the data into a Pandas DataFrame.


```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df['target'] = iris.target

df['species'] = df['target'].apply(lambda x: iris.target_names[x])

print(df.head())
```

## Task 1: Line Plot

Plot a **line graph** of `sepal length (cm)` values for the first 50 samples.  
*Hint: Use `plt.plot`*


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

sepal_length = iris.data[:50, 0]   

x = range(1, 51)

plt.plot(x, sepal_length, marker='o', linestyle='-', color='b')

plt.title("Sepal Length (cm) - Ä°lk 50 nÃ¼munÉ™")
plt.xlabel("NÃ¼munÉ™ nÃ¶mrÉ™si")
plt.ylabel("Sepal length (cm)")

plt.show()
```

## Task 2: Scatter Plot

Create a **scatter plot** of `sepal length (cm)` vs `sepal width (cm)` colored by species.  
*Hint: Use `plt.scatter`*


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

for i, target_name in enumerate(target_names):
    plt.scatter(
        X[y == i, 0],  
        X[y == i, 1],  
        label=target_name
    )

plt.title("Sepal Length vs Sepal Width (cm)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()

plt.show()
```

## Task 3: Bar Chart

Create a **bar chart** of the average `petal length (cm)` for each species.  
*Hint: Use `plt.bar`*


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

avg_petal_length = []
for i in range(len(target_names)):
    avg_petal_length.append(X[y == i, 2].mean())  

plt.bar(target_names, avg_petal_length, color=['skyblue', 'orange', 'green'])


plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")


plt.show()
```

## Task 4: Histogram

Plot a **histogram** of `petal width (cm)` values.  
*Hint: Use `plt.hist`*


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data

petal_width = X[:, 3]

plt.hist(petal_width, bins=20, color="purple", edgecolor="black", alpha=0.7)

plt.title("Histogram of Petal Width (cm)")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")


plt.show()
```

## Task 6: Pie Chart

Create a **pie chart** showing the proportion of each species.  
*Hint: Use `plt.pie`*


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Dataset yÃ¼klÉ™nmÉ™si
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

# Species adlarÄ±nÄ± É™lavÉ™ edirik (0,1,2 É™vÉ™zinÉ™ adlar yazÄ±lsÄ±n)
df["species_name"] = df["species"].map({0:"setosa", 1:"versicolor", 2:"virginica"})

# HÉ™r nÃ¶v Ã¼Ã§Ã¼n sayÄ±nÄ± hesablayÄ±rÄ±q
species_counts = df["species_name"].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    species_counts, 
    labels=species_counts.index, 
    autopct='%1.1f%%',     # Faiz gÃ¶stÉ™ricisi
    startangle=90,         # BaÅŸlama bucaÄŸÄ±
    colors=["skyblue", "lightgreen", "salmon"]
)
plt.title("Proportion of Iris Species")
plt.show()
```

## Task 7: Area Plot

Plot an **area plot** of the first 30 samples of sepal length and petal length.  
*Hint: Use `plt.fill_between`*


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Dataset yÃ¼klÉ™nmÉ™si
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Ä°lk 30 nÃ¼munÉ™nin sepal length vÉ™ petal length dÉ™yÉ™rlÉ™rini gÃ¶tÃ¼rÉ™k
x = range(30)  # nÃ¼munÉ™lÉ™rin indeksi
sepal_length = df['sepal length (cm)'][:30]
petal_length = df['petal length (cm)'][:30]

# Area plot (sahÉ™ qrafiki)
plt.figure(figsize=(10,6))
plt.fill_between(x, sepal_length, color="skyblue", alpha=0.5, label="Sepal Length")
plt.fill_between(x, petal_length, color="orange", alpha=0.5, label="Petal Length")

plt.title("Area Plot - First 30 Samples")
plt.xlabel("Sample Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()
```

## Task 8: Subplots

Create **subplots** with 2 rows and 2 columns showing:
1. Line plot of sepal length (first 50 samples)
2. Scatter plot of sepal length vs sepal width
3. Histogram of petal width
4. Box plot of features

*Hint: Use `plt.subplot`*


```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Dataseti yÃ¼klÉ™yirik vÉ™ DataFrame ÅŸÉ™klinÉ™ salÄ±rÄ±q
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

plt.figure(figsize=(12, 10))  # ÅžÉ™kilin Ã¼mumi Ã¶lÃ§Ã¼sÃ¼

# 1. Line plot (ilk 50 nÃ¼munÉ™nin sepal length dÉ™yÉ™rlÉ™ri)
plt.subplot(2, 2, 1)
plt.plot(df['sepal length (cm)'][:50], marker="o")
plt.title("Line Plot - Sepal Length (first 50 samples)")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")

# 2. Scatter plot (sepal length vs sepal width, nÃ¶vlÉ™rÉ™ gÃ¶rÉ™ rÉ™nglÉ™ndirilmiÅŸ)
plt.subplot(2, 2, 2)
for species, color in zip(df['species'].unique(), ["red", "green", "blue"]):
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'],
                label=species, alpha=0.7, color=color)
plt.title("Scatter Plot - Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()

# 3. Histogram (petal width)
plt.subplot(2, 2, 3)
plt.hist(df['petal width (cm)'], bins=20, color="purple", edgecolor="black")
plt.title("Histogram - Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")

# 4. Box plot (bÃ¼tÃ¼n feature-lÉ™r)
plt.subplot(2, 2, 4)
plt.boxplot(df.iloc[:, 0:4].values, labels=df.columns[0:4])
plt.title("Box Plot - Features")

# Layout dÃ¼zÉ™ltmÉ™k
plt.tight_layout()
plt.show()
```
